```python
#Pilot Navigation System

# You are the pilot of a plane, you need to land it in the upcoming airport
# For landing the plane you need to attain the Altitude of 3000ft, which is safe for landing
# if your plane is above 3000ft and below 6000ft, atleast allow the pilot to try for landing "Come down to 3000ft and open landing gear::"
# If your plane is above 6000ft, dnt allow the pilot to land the plane and ask "Go around"
# The user will enter the altitude of the plane
```

```python
pilot = int(input("Enter the altitude to land the plane:"))
altitude = int(pilot)
#lst = ["3000","6000"]
if(altitude == 3000):
    print("Safe for landing")
elif (altitude>3000 or altitude<6000):
    print("Come down to 3000ft and open landing gear")
elif altitude>6000:
    print("Don't land the plane, Go around")
else:
  print("Provided user Altitude not matching required altitude to land the plane")
```
